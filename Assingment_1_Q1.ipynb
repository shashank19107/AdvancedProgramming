```python
# Defining data sources
source = "https://archive.ics.uci.edu/ml/machine-learning-databases/abalone/abalone.data"
```

```python
df = pd.read_csv(source, sep = ",", index_col=False)
df.columns = ["Sex", "Length", "Diameter", "Height", "Whole weight", "Shucked weight", "Viscera weight", "shell weight", "Rings"]
pd.get_dummies(df.Sex)
```

Here we are creating a one-hot encoding for the variable sex, because its a descrete variable and we beleive that one-encoding best represents the information it carries.

```python
one_hot_code_sex = pd.get_dummies(df.Sex, prefix="Sex")
df = pd.concat([df, one_hot_code_sex], axis=1)
df = df.drop(columns="Sex")
df
```
